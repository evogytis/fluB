```python
#from IPython.external import mathjax; mathjax.install_mathjax()
import pandas as pd
from pandas import Series, DataFrame, Panel
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
from datetime import datetime, date, time
import sys
import os
import glob
import re
import csv
import Tkinter, tkFileDialog
import scipy as sp
from scipy.stats.stats import pearsonr
from scipy.stats.stats import spearmanr

import platform
import IPython

if 'Win' in platform.platform():
    splitter='\\'
    print 'windows. much improve. so amaze. wow'
else:
    splitter='/'
    print 'mac. much smarm. so impress. wow'
    
print 'OS:\t%s\npython version:\t%s\nIPython version:\t%s\nnumpy version:\t%s\nmatplotlib version:\t%s\npandas version:\t%s'%(platform.platform(),sys.version,'.'.join(map(str,IPython.version_info)[:-1]),np.__version__,mpl.__version__,pd.__version__)

def index(data,item):
    return [i for i,x in enumerate(data) if x == item]

def index2(data,item):
    return [i for i,x in enumerate(data) if item in x]

def column(data,col):
    return [row [col] for row in data]

def unique(o, idfun=repr):
    seen = {}
    return [seen.setdefault(idfun(e),e) for e in o if idfun(e) not in seen]

def interpretDate(date):
    out=0
    try:
        yrs = int(date.split('-')[0])
        mons = int(date.split('-')[1])
        dds = int(date.split('-')[2])
        out=float(float(dt(yrs,mons,dds).strftime("%j"))-1) / 366 + float(dt(yrs,mons,dds).strftime("%Y"))
    except IndexError:
        try:
            u = int(calendar.monthrange(int(date.split('-')[0]),int(date.split('-')[1]))[1])
            l = 1
            yrs = int(date.split('-')[0])
            mons = int(date.split('-')[1])
            dds=int((u+l)/float(2))
            out=float(float(dt(yrs,mons,dds).strftime("%j"))-1) / 366 + float(dt(yrs,mons,dds).strftime("%Y"))
        except IndexError:
            yrs = int(date)
            mons=1
            dds=1
            out=float(date)

    return out

def hpd(data, level):
    """
    Return highest posterior density interval from a list,
    given the percent posterior density interval required.
    """
    d = list(data)
    d.sort()

    nData = len(data)
    nIn = int(round(level * nData))
    if nIn < 2 :
        return None
   #raise RuntimeError("Not enough data. N data: %s"%(len(data)))
 
    i = 0
    r = d[i+nIn-1] - d[i]
    for k in range(len(d) - (nIn - 1)) :
        rk = d[k+nIn-1] - d[k]
        if rk < r :
            r = rk
            i = k

    assert 0 <= i <= i+nIn-1 < len(d)
 
    return (d[i], d[i+nIn-1])

def ESS(data, stepSize = 1) :
    samples = len(data)
    assert len(data) > 1,"no stats for short sequences"

    maxLag = min(samples//3, 1000)
    gammaStat = [0,]*maxLag
    varStat = 0.0;

    if type(data) != np.ndarray :
        data = np.array(data)

    normalizedData = data - data.mean()

    for lag in range(maxLag) :
        v1 = normalizedData[:samples-lag]
        v2 = normalizedData[lag:]
        v = v1 * v2
        gammaStat[lag] = sum(v) / len(v)
        if lag == 0 :
            varStat = gammaStat[0]
        elif lag % 2 == 0 :
            s = gammaStat[lag-1] + gammaStat[lag]
            if s > 0 :
                varStat += 2.0*s
            else:
                break

    act = stepSize * varStat / gammaStat[0]
    ess = (stepSize * samples) / act

    return ess

import collections
def overlap(a,b):
    """
    Return the elements shared by two lists in the following format:
    [overlap],[list 1 remainder],[list 2 remainder]
    """
    a_multiset = collections.Counter(a)
    b_multiset = collections.Counter(b)

    overlap = list((a_multiset & b_multiset).elements())
    a_remainder = list((a_multiset - b_multiset).elements())
    b_remainder = list((b_multiset - a_multiset).elements())

    return overlap, a_remainder, b_remainder

def getOverlap(a, b):
    """
    Return overlap between two lines given as [a1,a2],[b1,b2].
    """
    return max(0, min(a[1], b[1]) - max(a[0], b[0]))

def matrixIt(col1,col2,col3,filler,symmetry):
    """
    Returns a matrix from pairwise comparisons.
    Input format is (rows,columns,values,missing value indicator).
    Output is a list of lists.
    """
    rows=map(str,(sorted(unique(col1))))
    cols=map(str,(sorted(unique(col2))))
    fil=str(filler)
    outMatrix=list(list(filler for u in range(len(cols)+1)) for t in range(len(rows)+1))
    outMatrix[0][0]=''
    for i in range(len(outMatrix)-1):
        outMatrix[i+1][0]=str(rows[i])
    for j in range(len(outMatrix[0])-1):
        outMatrix[0][j+1]=str(cols[j])
    
    for v in range(len(col3)):
        try:
            rIndex=index(column(outMatrix,0),str(col1[v]))[0]
            cIndex=index(outMatrix[0],str(col2[v]))[0]
            outMatrix[rIndex][cIndex]=str(col3[v])
            if symmetry==True:
                outMatrix[cIndex][rIndex]=str(col3[v])
        except:
            print rIndex,cIndex,col1[v],col2[v],col3[v]
        
    return outMatrix

def readMatrix(matrix):
    matrixForm=[]
    cols=[]
    rows=[]
    counter=0
    for entry in matrix:
        if counter==0:
            counter+=1
            cols=entry[1:]
        else:
            matrixForm.append(entry[1:])
            rows.append(entry[0])
            
    return matrixForm,rows,cols

def adjustTickLabels(figureObject,axisObject,which,indices):
    """
    Given fig, ax, x or y and indices replaces them with empty strings
    """
    figureObject.canvas.draw()
    if which=='x':
        labels=[item.get_text() for item in ax.get_xticklabels()]
    else:
        labels=[item.get_text() for item in ax.get_yticklabels()]
       
    for value in indices:
        labels[value]=''
    
    if which=='x':
        ax.set_xticklabels(labels)
    else:
        ax.set_yticklabels(labels)

def blend_palette(colors, n_colors=6, as_cmap=False):
    """Make a palette that blends between a list of colors.

Parameters
----------
colors : sequence of matplotlib colors
hex, rgb-tuple, or html color name
n_colors : int, optional
number of colors in the palette
as_cmap : bool, optional
if True, return as a matplotlib colormap instead of list

Returns
-------
palette : list or colormap

"""
    name = "-".join(map(str, colors))
    pal = mpl.colors.LinearSegmentedColormap.from_list(name, colors)
    if not as_cmap:
        pal = pal(np.linspace(0, 1, n_colors))
    return pal

def treeName(filename):
    if filename=='M1':
        return 'MP'
    elif filename=='NS1':
        return 'NS'
    else:
        return filename

def palplot(pal, size=1):
    """Plot the values in a color palette as a horizontal array.

Parameters
----------
pal : sequence of matplotlib colors
colors, i.e. as returned by seaborn.color_palette()
size :
scaling factor for size of plot

"""
    n = len(pal)
    f, ax = plt.subplots(1, 1, figsize=(n * size, size))
    ax.imshow(np.arange(n).reshape(1, n),
              cmap=mpl.colors.ListedColormap(list(pal)),
              interpolation="nearest", aspect="auto")
    ax.set_xticks(np.arange(n) - .5)
    ax.set_yticks([-.5, .5])
    ax.set_xticklabels([])
    ax.set_yticklabels([])

PB1=(0.2981622478541206, 0.44644367598435458, 0.64719724652229571)
PB2=(0.28098424626331703, 0.63955403192370541, 0.39507882933990629)
PA=(0.8162552973803352, 0.41062668737243202, 0.29078047883276847)
HA=(0.88366013765335083, 0.77908497055371606, 0.18562091886997223)
NP=(0.8, 0.4745098039215686, 0.6549019607843137)
NA=(0.4666666666666667, 0.7450980392156863, 0.8588235294117647)
MP=(0.49803923567136127, 0.43267975250879925, 0.52287583549817396)
NS=(0.8, 0.7552941176470589, 0.5929411764705883)
colours=[PB1,PB2,PA,HA,NP,NA,MP,NS]

palplot(colours)

print '\nEverything loaded!'
```
